# The Big Move: Working out where to live in Toronto

## Collecting Toronto neighbourhood data with coordinates

```python
import pandas as pd
import numpy as np
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import folium
import requests
from pandas.io.json import json_normalize
```

```python
# Scraping & storing postal code data
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

webpage = pd.read_html(url)

df = webpage[0]
```

```python
# Dropping postal codes that do not have an assigned borough
df = df[df.Borough != 'Not assigned']

# Sorting results & resetting the index
df = df.sort_values(['Postal Code'])
df = df.reset_index(drop = True)

# Checking if any cells have a borough but a Not Assigned neighbourhood
df[df['Neighbourhood'] == 'Not assigned']
```

```python
# Reading geospatial coordinates data
coord = pd.read_csv('http://cocl.us/Geospatial_data')

# Combining datasets
df_geo = pd.merge(df, coord)
print(df_geo.shape)
df_geo.head()
```

```python
# Calculating distance between each neighbourhood & Toronto Union Station using Haversine formula
# (https://latitude.to/articles-by-country/ca/canada/7647/union-station-toronto)

# Radius of the earth
R = 6373.0

# Setting coordinates (lat1/lon1 for Union Station, lat2/lon2 for each neighbourhood)
lat1 = []
lon1 = []
lat2 = []
lon2 = []

for i in [43.640497438] * 103:
    lat1.append(math.radians(i))

for i in [-79.3749985] * 103:
    lon1.append(math.radians(i))
    
for i in df_geo['Latitude']:
    lat2.append(math.radians(i))

for i in df_geo['Longitude']:
    lon2.append(math.radians(i))

# Defining change in coordinates
dlat = []
dlon = []

for i in range(103):
    dlat.append(lat2[i] - lat1[i])

for i in range(103):
    dlon.append(lon2[i] - lon1[i])

# Haversine formula
a = []
c = []
distance = []

for i in range(103):
    a.append(math.sin(dlat[i] / 2)**2 + math.cos(lat1[i]) * math.cos(lat2[i]) * math.sin(dlon[i] / 2)**2)
    c.append(2 * math.atan2(math.sqrt(a[i]), math.sqrt(1 - a[i])))
    distance.append(c[i] * R)
    
# Appending distance information to dataset
df_geo['Distance'] = distance
df_geo.head()
```

```python
# Calculating distance between North Melbourne & Flinders Street Station using Haversine formula
# (https://latitude.to/articles-by-country/au/australia/39327/north-melbourne-victoria)
# (https://latitude.to/articles-by-country/au/australia/13303/flinders-street-railway-station)

R = 6373.0

lat_nm = math.radians(-37.792830162)
lon_nm = math.radians(144.941329568)

lat_fss = math.radians(-37.818078)
lon_fss = math.radians(144.96681)

d_lat = lat_nm - lat_fss
d_lon = lon_nm - lon_fss

a_nm = math.sin(d_lat / 2)**2 + math.cos(lat_fss) * math.cos(lat_nm) * math.sin(d_lon / 2)**2
c_nm = 2 * math.atan2(math.sqrt(a_nm), math.sqrt(1 - a_nm))
distance_nm = c_nm * R

# So the distance between North Melbourne & Flinders Street Station is:
print(distance_nm)
```

```python
# Assuming a willingness to live within double the distance (from Toronto city centre) of North Melbourne from Melbourne city centre:
df_close = df_geo[df_geo['Distance'] < 2 * distance_nm]

print(df_close.shape)
df_close.tail()
```

```python
# Appending North Melbourne to dataset
df_comb = df_close
df_comb.loc[32] = ["3051", "City of Melbourne", "North Melbourne", -37.792830162, 144.941329568, distance_nm]
df_comb.tail()
```

```python
# Defining function to get nearby venues in each neighbourhood
LIMIT = 100

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Running the above function for Toronto neighbourhoods & North Melbourne
toronto_venues = getNearbyVenues(names = df_comb['Neighbourhood'],
                                 latitudes = df_comb['Latitude'],
                                 longitudes = df_comb['Longitude']
                                )
```

```python
print(toronto_venues.shape)
toronto_venues.head()
```

```python
# Grouping venues by neighbourhood
toronto_venues.groupby('Neighbourhood').count()
```

```python
# Breaking down venues by category
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# Add neighbourhood column back to dataframe
toronto_onehot['Neighbourhood'] = toronto_venues['Neighbourhood'] 

# Move neighbourhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

# Grouping by neighbourhood - mean
toronto_grouped = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
toronto_grouped
```

```python
# Top 5 most common venue categories by neighbourhood
num_top_venues = 5

for hood in toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# Storing most common venue categories by neighbourhood
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# Creating columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# Creating a new dataframe
neighbourhoods_venues_sorted = pd.DataFrame(columns=columns)
neighbourhoods_venues_sorted['Neighbourhood'] = toronto_grouped['Neighbourhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighbourhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

print(neighbourhoods_venues_sorted.shape)
neighbourhoods_venues_sorted.head()
```

```python
# Clustering the neighbourhoods with different k values
toronto_grouped_clustering = toronto_grouped.drop('Neighbourhood', 1)

km_inertia = []

for i in range(1,11):
    kclusters = i
    kmeans = KMeans(init = "k-means++", n_clusters=kclusters, n_init=10).fit(toronto_grouped_clustering)
    km_inertia.append(kmeans.inertia_)

# Plotting the distortion function for each k value
mpl.pyplot.plot(range(1,11), km_inertia)
```

```python
# Taking k = 4 as the elbow point
kclusters = 4
kmeans = KMeans(init = "k-means++", n_clusters=kclusters, n_init=10).fit(toronto_grouped_clustering)
kmeans.labels_
```

```python
# Adding cluster labels to the dataframe
kmeans_labels = kmeans.labels_.tolist()

labels = {'Neighbourhood':toronto_grouped[['Neighbourhood']],'Cluster Labels':kmeans_labels}

df_labels = pd.DataFrame(toronto_grouped[['Neighbourhood']])
df_labels['Cluster Labels'] = kmeans_labels

# Add cluster labels for each neighbourhood
toronto_labelled = pd.merge(df_labels, neighbourhoods_venues_sorted)

# Add coordinates for each neighbourhood
toronto_merged = pd.merge(df_comb, toronto_labelled)

toronto_merged.head()
```

```python
# Check neighbourhoods in the same cluster as North Melbourne (Cluster 1)
toronto_similar = toronto_merged[toronto_merged['Cluster Labels'] == 1]
toronto_similar = toronto_similar[toronto_similar['Neighbourhood'] != 'North Melbourne']
toronto_similar.shape
```
